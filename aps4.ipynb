___



Um coeficiente de convecção é uma medida que descreve a taxa de transferência de calor por convecção em um sistema. 
Pode-se mudar o coeficiente físicamente modificando fatores como a velocidade do fluido, e as características geométricas (superfície de contato) do sistema. Quanto maior o seu valor, mais eficiente é a transferência de calor por convecção.

___






____



As aletas contribuíram de forma a aumentar a superfície de contato entre o material que compõe a placa com o fluido da vizinhança, consequentemente facilitando a troca de calor entre o processador e o ambiente, não só diminuindo sua temperatura, mas também deixando a distribuição dessa temperatura mais homogênea.

___



Ao mudar a temperatura ambiente, mudamos a temperatura do fluido de vizinhança que envolve a face exposta do processador, isso não muda exatamente a forma com a qual a temperatura é distribuida, mas sim os valores absolutos dessa temperatura nas regiões da placa.
